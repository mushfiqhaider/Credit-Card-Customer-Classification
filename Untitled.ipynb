```python
df['Attrition_Flag']=df["Attrition_Flag"].replace({'Existing Customer':1,'Attrited Customer':0})
df['Gender']=df['Gender'].replace({'M':1,'F':0})
df['Education_Level']=df['Education_Level'].replace({'Graduate':0,'High School':1,'Unknown':3,'Uneducated'
                                                    :4,'College':5,'Post-Graduate':6,'Doctorate':7})
df['Marital_Status']=df['Marital_Status'].replace({'Married':0,'Single':1,'Unknown':2,'Divorced':3})
df['Income_Category']=df['Income_Category'].replace({'Less than $40K':0,'$40K - $60K':1,'$80K - $120K':2
                                                    ,'$60K - $80K':3,'Unknown':4,'$120K +':5})
df['Card_Category']=df['Card_Category'].replace({'Blue':0,'Silver':1,'Gold':2,'Platinum':3})
```

```python
df.info()
```

```python
x=df.drop('Attrition_Flag',axis=1)
y=df['Attrition_Flag']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg=RandomForestClassifier()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
